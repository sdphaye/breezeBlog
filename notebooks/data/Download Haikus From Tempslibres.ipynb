```python
import time
!pip install langdetect
from bs4 import BeautifulSoup
from langdetect import detect
import pandas as pd
import requests
```

```python
i = 0
haikus = []
```

```python
while True:
    time.sleep(.2)
    response = requests.get('https://www.tempslibres.org/tl/tlphp/dbauteursl.php?lang=en&lg=e' % i)
    soup = BeautifulSoup(response.text, 'html.parser')
    lines = soup.find('p', 'dbhkanahaiku').get_text().splitlines()
    if lines:
        if '' in lines:
            haiku = lines[:lines.index('')]
        else:
            haiku = lines

        print(i, haiku)
        haikus.append(haiku)
        
    i += 1
```

```python
haikus2 = haikus[:7599]
```

```python
haikus3 = []
for haiku in haikus2:
    try:
        haikus3.append([line.encode('latin1').decode('utf-8') for line in haiku])
    except:
        print(haiku)
```

```python
df = pd.DataFrame.from_records(haikus3)
df['haiku'] = df.apply(lambda r: '\n'.join([l for l in r[0:9] if l]), axis=1)
```

```python
df
```

```python
df['lang'] = df['haiku'].apply(lambda h: detect(h))
```

```python
df.to_csv('../../input/poems/raw/tempslibres.csv')
```
